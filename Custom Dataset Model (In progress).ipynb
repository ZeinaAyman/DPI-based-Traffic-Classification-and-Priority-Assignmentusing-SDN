```python
import pandas as pd 
```

```python
dataset = pd.read_csv("d:/dataset.csv" , low_memory=False) 
df = pd.DataFrame(dataset)
display(df)
print (df['priority'].value_counts(ascending=True))
print("Dataset size: " ,df.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
enc.fit(df['Day'])
df['Day'] = enc.transform(df['Day'])
print(df['Day'].values)
#days_mapping={'Sat':0,'Sun':1,'Mon':2,'Tue':3,'Wed':4,'Thu':5,'Fri':6}
#df['Day']=df['Day'].replace(days_mapping)
```

```python
import numpy as np
df['day_of_week_sin'] = np.sin(df['Day'] * (2 * np.pi / 7))
df['day_of_week_cos'] = np.cos(df['Day'] * (2 * np.pi / 7))

print(df['day_of_week_cos'])
display(df)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer 
from sklearn.feature_extraction.text import CountVectorizer 

x=df['Keywords']      

print(df['Keywords'])

x=CountVectorizer().fit_transform(x.apply(lambda x: np.str_(x)))
   
x=TfidfTransformer().fit_transform(x) 

df['Keywords']=x

```

```python
df = df[['ID', 'Day', 'day_of_week_sin','day_of_week_cos','Start Time','End Time','URL','Keywords','type','content','priority']]
```

```python
display(df)
```

```python

from datetime import datetime

i=0
for x in df['Start Time'].values:
  x = datetime.strptime(x, '%H:%M')
  df['Start Time'].values[i]=x
  i=i+1
```

```python
list=[]
for h in df['Start Time'].values:
    s=h.hour *60*60+ h.minute *60 
    list.append(s)
    #df['sin_time']=sin(2*np.pi*sec/seconds_in_day)

df['Start Time']=list
```

```python

seconds_in_day = 24*60*60


df['start_time_sin'] = np.sin(df['Start Time'] * (2 * np.pi / seconds_in_day))
df['start_time_cos'] = np.cos(df['Start Time'] * (2 * np.pi / seconds_in_day))

print(df['start_time_sin'].values)
print(df['start_time_sin'].values)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelBinarizer

jobs_encoder = LabelBinarizer()
jobs_encoder.fit(df['type'])
transformed = jobs_encoder.transform(df['type'])
ohe_df = pd.DataFrame(transformed)
df = pd.concat([df, ohe_df], axis=1).drop(['type'], axis=1)
```

```python
jobs_encoder = LabelBinarizer()
jobs_encoder.fit(df['content'])
transformed = jobs_encoder.transform(df['content'])
ohe_df = pd.DataFrame(transformed)
df = pd.concat([df, ohe_df], axis=1).drop(['content'], axis=1)
```

```python
jobs_encoder = LabelBinarizer()
jobs_encoder.fit(df['URL'])
transformed = jobs_encoder.transform(df['URL'])
ohe_df = pd.DataFrame(transformed)
df = pd.concat([df, ohe_df], axis=1).drop(['URL'], axis=1)
```

```python
display(df)
```

```python
df1 = df.pop('priority') # remove column b and store it in df1
df['priority']=df1 # add b series as a 'new' column.

display(df)
```

```python
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
#df = df.drop('End Time', 1)
#df = df.drop('Keywords', 1)

df.replace([np.inf, -np.inf], np.nan, inplace=True) #replace infinity values with NaN
df.dropna(inplace=True) #dropping rows with missing values  

count=df.shape[1]
x = df.iloc[:,:-1].values
y = df.iloc[:, count-1].values
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

knn = KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))
```
